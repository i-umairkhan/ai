# Matplotlib Cheatsheet

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# x axis data
x = np.arange(0,11)
x
```

```python
# y axis data
y = x * x
y
```

```python
# creating graph using plot
plt.plot(x,y)
#  axis labels
plt.xlabel("Days") 
plt.ylabel("Infection Rates")
plt.title("Infection Rates Over Times")
# plt.savefig("graph.png") -> save image of graph
```

```python
# graph using figure 
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel("Days")
axes.set_ylabel("Infection Rates")
axes.set_title("Infection Rates Over Time")
```

```python
# graph using figure with multiple axes
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel("Days")
axes.set_ylabel("Infection Rates")
axes.set_title("Infection Rates Over Time")

axes2 = fig.add_axes([0.25,0.5,0.25,0.25])
axes2.plot(x,y)
# limiting graph view points
axes2.set_xlim(3,5)
axes2.set_ylim(10,30)
axes2.set_title("Critical Point")
```

```python
# data
a = np.arange(0,11)
b = a * 3
```

```python
# creating graph with multiple data and labels
fig = plt.figure(dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,label="Bactaria")
axes.set_xlabel("Days")
axes.set_ylabel("Infection Rates")
axes.set_title("Infection Rates Over Time")
axes.plot(a,b,label="Virus")
axes.legend()
```

```python
# creating subplots
fig, axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
fig.subplots_adjust(wspace=0.25,hspace=0.5)
```
