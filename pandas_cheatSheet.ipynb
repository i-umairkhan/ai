## pandas meathods

```python
df = pd.read_csv('cars.csv')
```

```python
def price_after(price):
    return round(price *1.13,0)
# apply meathod takes a coloumn and applyes a function on every element of coloumn
# apply meathod can only be on one coloumn
df["total_price"] = df["price"].apply(price_after)
df.head()
```

```python
# this fuunction takes price and finance and returns status based on it
def profit_margin(price,finance):
    if finance == 'N':
        if price < 10000:
            return "LOW"
        elif price >= 10000 and price < 25000:
            return "MED"
        else:
            return "HIGH"
    else:
        if price < 7500:
            return "LOW"
        elif price >= 7500 and price < 20000:
            return "MED"
        else:
            return "HIGH"
```

```python
# this vectorize functions comes from np it does same task but can be apllied to multiple columns
df ['profit_margin'] = np.vectorize(profit_margin)(df["price"],df['finance'])
df.head()
```
